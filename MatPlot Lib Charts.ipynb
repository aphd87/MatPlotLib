```python
#Import mat plot library 
import matplotlib.pyplot as plt

import numpy as np
```

```python
adv_budget = [80, 100, 120, 150,170]
sales = [45, 50, 55, 60, 80]
```

```python
#Line chart 
plt.plot(adv_budget, sales, 'r')       #r is the color red
plt.xlabel('Adv Budget')              # x axis label
plt.ylabel('Sales of Product')        #y axis label
plt.title('Impact of Adv on Sales')   #title of graph
plt.show()                            #shows the graph
```
